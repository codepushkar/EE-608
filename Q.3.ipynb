```python
import numpy as np
x = np.array([1.1,2,3.3,4.2,4.9,6.1,7,8.3,9.1])
y = np.array([1.3,1.8,3,4.1,5.2,6.3,7,8.1,9.2])
```

```python
print("xi\tf(xi)")
for i,j in zip(x,y):
    print(f"{i}\t{j}")
```

```python
#Number of Maximum Iterations
max_iter = 1000

#Initialization of Parameters(Random Values)
m = 0.5
c = 0.5

#Rate of Learning
alpha = 0.001

losses = []
directions = []


best_m = 0
best_c = 0
best_loss = 0


for i in range(max_iter):
#Slope Eqn
    yi = m * x + c
    
    loss = 1/(2*len(y)) * sum((y - yi)**2)
    losses.append(loss)
    
    
    if loss == 0.001:
            print('Loss Pretty Low stopping.')
            
    if i != 0 and loss < losses[-2]:
        best_loss = loss
        best_m = m
        best_c = c

    if i % 100 == 0:
    
        print(f'for value {i + 1} loss {loss}')

        
    dm = (1/len(y)) * sum((y - yi) * -x)
    dc = (1/len(y)) * sum((y-yi))*-1
#Values of 10 Iterations
    if i < 10:
        directions.append((dm,dc))

    m = m - alpha * dm
    c = c - alpha * dc  
```

```python
import matplotlib.pyplot as plt

iterations = list(range(max_iter))

plt.plot(iterations,losses)
plt.xlabel("Iteration Number")
plt.ylabel("Objective Function")
```

```python
#Computing Steepest Descent Directions (Iteration = 10)
for i,(m_,c_) in enumerate(directions):
    print(f"step {i} m = {m_:.4f} and c = {c_:.4f}")
```

```python
dir = np.array(directions)

m_ = dir[:,0]
c_ = dir[:,1]

plt.plot(m_,label = 'c')
plt.plot(c_,label = 'm')
plt.legend()
```

```python
#Optimal Parameter Values
print(f"Best m = {best_m:.4f} Best c = {best_c:.4f} : loss = {best_loss:.4f}")
```
